```python
#random dataset generation

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlt

mlt.rcParams.update({'font.size':14,'figure.dpi':100})
rng=np.random.default_rng(44)

beta1=4.0
beta0=2.0

n=140
x=rng.uniform(0,10,size=n)
noise=rng.normal(0,2,size=n)
y=beta1*x+beta0+noise

X=np.column_stack([np.ones(n),x])

print(y.shape)
print(X.shape)

plt.figure(figsize=(7, 4))
plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.4, label='data')
plt.xlabel('x'); plt.ylabel('y')
plt.title('Training data')
plt.legend(); plt.tight_layout(); plt.show()
```

```python
#Linear regression class
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlt
class linearRegressionNForm():
    def __init__(self):
        self.betaList=None
    
    def fit(self,x,y):
        covmatrix=x.T @ x
        yxtr=x.T @ y
        self.betaList=np.linalg.solve(covmatrix,yxtr)
        return self
    
    def predict(self,x):
        return x @ self.betaList
    
    def mse(self,y,x):
        residualmat=y- x @ self.betaList
        return np.mean(residualmat**2)
        
#multivariate model prediction

X2 = np.array([
    [1, 2, 5, 7, 1],
    [1, 3, 8, 2, 4],
    [1, 4, 1, 9, 5],
    [1, 5, 3, 4, 8],
    [1, 6, 7, 6, 2],

    [1, 7, 2, 8, 9],
    [1, 8, 9, 3, 6],
    [1, 9, 4, 5, 7],
    [1, 10, 6, 1, 3],
    [1, 11, 8, 7, 4],

    [1, 12, 5, 2, 6],
    [1, 13, 7, 9, 8],
    [1, 14, 1, 4, 5],
    [1, 15, 3, 6, 7],
    [1, 16, 9, 8, 1],

    [1, 17, 2, 5, 3],
    [1, 18, 4, 7, 9],
    [1, 19, 6, 3, 2],
    [1, 20, 8, 1, 4],
    [1, 21, 5, 9, 6]

])
Y = np.array([

    58, 68, 68, 81, 74,
    93, 91, 88, 61, 89,
    76, 113, 72, 93, 100,
    82, 117, 88, 77, 110

])
model = linearRegressionNForm()
model.fit(X2,Y)
print(model.betaList)
print(model.mse(Y,X2))

# random dataset
mlt.rcParams.update({'font.size':14,'figure.dpi':100})
rng=np.random.default_rng(44)
beta1=4.0
beta0=2.0

n=140
x=rng.uniform(0,10,size=n)
noise=rng.normal(0,2,size=n)
y=beta1*x+beta0+noise

X=np.column_stack([np.ones(n),x])
model.fit(X,y)
print(model.betaList)

fig, axes= plt.subplots(1,2,figsize=(13,14))

#for left plot
a=axes[0]
y=model.predict(X)
a.plot(x,y)

# Right plot
ax = axes[1]
residuals = y - model.predict(X)
ax.scatter(model.predict(X), residuals, alpha=0.5, edgecolors='k', linewidths=0.4)
ax.axhline(0, color='r', lw=1.5, ls='--')
ax.set_xlabel('Predicted $\\hat{y}$'); ax.set_ylabel('Residual $y - \\hat{y}$')
ax.set_title('Residual plot (should look like noise)')

plt.show()
```
